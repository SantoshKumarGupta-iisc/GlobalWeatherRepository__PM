```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("GlobalWeatherRepository.csv")
```

```python
# Display basic information about the dataset
print("Initial Dataset Information:\n", df.info())
print("\nMissing Values Before Handling:\n", df.isnull().sum())
```

```python
# Handle missing values
# Filling numerical columns with median and categorical columns with mode
for column in df.columns:
    if df[column].dtype == 'object':  # Categorical column
        df[column] = df[column].fillna(df[column].mode()[0])  # Assign back explicitly
    else:  # Numerical column
        df[column] = df[column].fillna(df[column].median())  # Assign back explicitly
```

```python
# Detect outliers using IQR method
numerical_columns = df.select_dtypes(include=[np.number]).columns
outlier_counts = {}
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

# Print the count of outliers in each column
print("\nOutlier Count in Each Numerical Column:\n", outlier_counts)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df[numerical_columns]), columns=numerical_columns)

# Plot boxplot for scaled data
plt.figure(figsize=(20, 6))
scaled_df.boxplot(rot=90)
plt.title("Boxplot of Scaled Numerical Features")
plt.show()
```

```python
# Detect and handle outliers using IQR method
numerical_columns = df.select_dtypes(include=[np.number]).columns
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
```

```python
# Normalize numerical columns using MinMaxScaler
scaler = MinMaxScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
# Display dataset after preprocessing
print("\nMissing Values After Handling:\n", df.isnull().sum())
print("\nDataset After Cleaning and Normalization:\n", df.head())
```

```python
# Save the cleaned dataset
df.to_csv("GlobalWeatherRepository_Cleaned.csv", index=False)
print("\nData Cleaning and Preprocessing Completed. Cleaned data saved as 'GlobalWeatherRepository_Cleaned.csv'.")
```

## Exploratory Data Analysis (EDA)

```python
# Load the cleaned dataset
df = pd.read_csv("GlobalWeatherRepository_Cleaned.csv")
```

```python
original_df = pd.read_csv("GlobalWeatherRepository.csv")  # Load original dataset
numerical_columns = original_df.select_dtypes(include=[np.number]).columns

min_max_values = pd.DataFrame({
    "column": numerical_columns,
    "min": original_df[numerical_columns].min(),
    "max": original_df[numerical_columns].max()
}).reset_index(drop=True)

min_max_values.to_csv("min_max_values.csv", index=False)  # Save min-max values
print("min_max_values.csv saved successfully.")
```

```python
# Denormalize numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns
original_min_max = pd.read_csv("min_max_values.csv")
for col in numerical_columns:
    min_val = original_min_max.loc[original_min_max['column'] == col, 'min'].values[0]
    max_val = original_min_max.loc[original_min_max['column'] == col, 'max'].values[0]
    df[col] = df[col] * (max_val - min_val) + min_val
```

# Advanced Assessment

```python
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
# Anomaly Detection using Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
df['anomaly'] = iso_forest.fit_predict(df[numerical_columns])

# Visualizing Anomalies
plt.figure(figsize=(10, 5))
anomalies = df[df['anomaly'] == -1]
sns.scatterplot(x=df.index, y=df['temperature_celsius'], label="Normal", color='blue')
sns.scatterplot(x=anomalies.index, y=anomalies['temperature_celsius'], label="Anomaly", color='red')
plt.title("Temperature Anomalies")
plt.xlabel("Index")
plt.ylabel("temperature_celsius")
plt.legend()
plt.show()

# Forecasting with ARIMA
train, test = train_test_split(df['temperature_celsius'], test_size=0.2, shuffle=False)
model_arima = ARIMA(train, order=(5,1,0))
model_arima_fit = model_arima.fit()
predictions_arima = model_arima_fit.forecast(steps=len(test))

# Evaluate ARIMA Model
mae_arima = mean_absolute_error(test, predictions_arima)
print("ARIMA MAE:", mae_arima)

# Forecasting with Moving Average
# Compute moving average
moving_avg = train.rolling(window=5).mean().dropna()

# Copy test data for predictions
predictions_ma = test.copy()

# Assign moving average values safely
min_length = min(len(predictions_ma), len(moving_avg))
predictions_ma.iloc[:min_length] = moving_avg.values[:min_length]

# Evaluate Moving Average Model
mae_ma = mean_absolute_error(test.iloc[:min_length], predictions_ma.iloc[:min_length])
print("Moving Average MAE:", mae_ma)


# Ensemble Model (Averaging ARIMA and Moving Average Predictions)
predictions_ensemble = (predictions_arima + predictions_ma) / 2
mae_ensemble = mean_absolute_error(test.iloc[:len(predictions_ensemble)], predictions_ensemble)
print("Ensemble Model MAE:", mae_ensemble)

print("Advanced EDA and Forecasting Completed.")
```

```python
from sklearn.ensemble import RandomForestRegressor
# List of air quality parameters
air_quality_targets = ['air_quality_Carbon_Monoxide', 'air_quality_Ozone', 'air_quality_PM2.5', 'air_quality_PM10']

# Initialize a dictionary to store feature importances
feature_importance_dict = {}

# Loop through each target air quality parameter
for target in air_quality_targets:
    # Drop rows with missing values in selected columns
    data = df[air_quality_columns + weather_columns].dropna()

    # Split into features and target
    X = data[weather_columns]
    y = data[target]

    # Train a Random Forest model
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X, y)

    # Get feature importances
    importance = pd.Series(model.feature_importances_, index=weather_columns).sort_values(ascending=False)
    feature_importance_dict[target] = importance

    # Plot feature importance
    plt.figure(figsize=(8, 5))
    sns.barplot(x=importance.values, y=importance.index, hue=importance.index, palette='viridis', legend=False)
    plt.title(f'Feature Importance for Predicting {target}')
    plt.xlabel('Importance')
    plt.ylabel('Weather Features')
    plt.show()

# Display the feature importance values for all targets
for target, importance in feature_importance_dict.items():
    print(f"\nFeature Importance for {target}:\n{importance}")
```
